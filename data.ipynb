### Store CSV into DataFrame

```python
# load and store dfs
citibike_2022_df = pd.concat(
    map(pd.read_csv, [
        'Resources/2022/JC-202201-citibike-tripdata.csv',
        'Resources/2022/JC-202202-citibike-tripdata.csv',
        'Resources/2022/JC-202203-citibike-tripdata.csv',
        'Resources/2022/JC-202204-citibike-tripdata.csv',
        'Resources/2022/JC-202205-citibike-tripdata.csv',
        'Resources/2022/JC-202206-citibike-tripdata.csv',
        'Resources/2022/JC-202207-citibike-tripdata.csv',
        'Resources/2022/JC-202208-citibike-tripdata.csv',
        'Resources/2022/JC-202209-citibike-tripdata.csv',
        'Resources/2022/JC-202210-citibike-tripdata.csv',
        'Resources/2022/JC-202211-citibike-tripdata.csv',
        'Resources/2022/JC-202212-citibike-tripdata.csv'
        ]), ignore_index=True)
# # output file path
# output_data_2022 = "./Resources/citibike_2022.csv"

citibike_2022_df.info()
citibike_2022_df
```

```python
# rename columns
citibike_2022_df = citibike_2022_df.rename(columns = {
    'ride_id': 'ride id',
    'rideable_type': 'ride type', 
    'started_at': 'start at', 
    'ended_at': 'end at',
    'start_station_name': 'start station name', 
    'start_station_id': 'start station id', 
    'end_station_name': 'end station name',
    'end_station_id': 'end station id', 
    'start_lat': 'start latitude', 
    'start_lng': 'start longitude', 
    'end_lat': 'end latitude', 
    'end_lng': 'end longitude',
    'member_casual': 'membership'
})
citibike_2022_df.head()
```

```python
#export to csv
citibike_2022_df.to_csv("./Resources/citibike_2022.csv", index = False)
```

```python
# load and store dfs
citibike_2021_df = pd.concat(
    map(pd.read_csv, [
        'Resources/2021/JC-202101-citibike-tripdata.csv',
        'Resources/2021/JC-202102-citibike-tripdata.csv',
        'Resources/2021/JC-202103-citibike-tripdata.csv',
        'Resources/2021/JC-202104-citibike-tripdata.csv',
        'Resources/2021/JC-202105-citibike-tripdata.csv',
        'Resources/2021/JC-202106-citibike-tripdata.csv',
        'Resources/2021/JC-202107-citibike-tripdata.csv',
        'Resources/2021/JC-202108-citibike-tripdata.csv',
        'Resources/2021/JC-202109-citibike-tripdata.csv',
        'Resources/2021/JC-202110-citibike-tripdata.csv',
        'Resources/2021/JC-202111-citibike-tripdata.csv',
        'Resources/2021/JC-202112-citibike-tripdata.csv'
        ]), ignore_index=True)
# output file path
output_data_2021 = "./Resources/citibike_2021.csv"
# #export to csv
# citibike_2021_df.to_csv("./Resources/citibike_2021.csv", index = False)

citibike_2021_df.info()
citibike_2021_df
```

```python
citibike_2021_df = citibike_2021_df.rename(columns = {
    'trip_duration': 'trip duration', 
    'started_at': 'start at', 
    'ended_at': 'end at', 
    'start_station_id': 'start station id',
    'start_station_name': 'start station name', 
    'start_lat': 'start latitude', 
    'start_lng': 'start longitude', 
    'end_station_id': 'end station id',
    'end_station_name': 'end station name', 
    'end_lat': 'end latitude', 
    'end_lng': 'end longitude', 
    'bike_id': 'bike id', 
    'member_casual': 'membership',
    'birth_year': 'birth year', 
    'gender': 'gender', 
    'ride_id': 'ride id', 
    'rideable_type': 'ride type'
    })
citibike_2021_df = citibike_2021_df.replace({
    'Customer': 'casual',
    'Subscriber': 'member'
})
citibike_2021_df['ride type'] = citibike_2021_df['ride type'].fillna('docked_bike')
citibike_2021_df
```

```python
#export to csv
citibike_2021_df.to_csv("./Resources/citibike_2021.csv", index = False)
```

```python
# stack 2021 csv on top of 2022 csv
citibike_2122_df = pd.concat(
    map(pd.read_csv, [
        './Resources/citibike_2021.csv',
        './Resources/citibike_2022.csv'
        ]), ignore_index=True)
# output file path
output_data_2122 = "./Resources/citibike_2122.csv"

citibike_2122_df.info()
citibike_2122_df
```
